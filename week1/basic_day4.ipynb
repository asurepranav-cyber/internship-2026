```python
# Get the number from the user
num = int(input("Enter a number: "))

# Variable to keep track of the running total
total_sum = 0

# Keep running as long as the number is greater than 0
while num > 0:
    # 1. Get the last digit
    last_digit = num % 10
    
    # 2. Add that digit to our total
    total_sum += last_digit
    
    # 3. Remove the last digit from the number (e.g., 123 // 10 = 12)
    num = num // 10

print(f"The sum of the digits is: {total_sum}")
#ex program 
```
