```python
"""
create a class called student
create 2-3 objects having attributes name, age

create a method which will add a skill (eg.. python,ML,Java) to the student 
where u can view the info of it later (i.e. skills learned by a student)
""";
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.skills = []

    # Method to add skill
    def add_skill(self, skill):
        self.skills.append(skill)

    # Method to display info
    def show_info(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Skills Learned:", ", ".join(self.skills))
        print()


# Creating objects
student1 = Student("Pranav", 20)
student2 = Student("Rahul", 21)
student3 = Student("Anjali", 19)

# Adding skills
student1.add_skill("Python")
student1.add_skill("Machine Learning")

student2.add_skill("Java")
student2.add_skill("C++")

student3.add_skill("Python")
student3.add_skill("Data Science")
```

```python
# Display student details
student1.show_info()
student2.show_info()
student3.show_info()
```

```python
class BankAccount:
    def __init__(self, name, balance=0):
        self.name = name
        self.balance = balance

    # Check balance
    def check_balance(self):
        print(f"\nCurrent Balance: ₹{self.balance}")

    # Deposit money
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"₹{amount} deposited successfully.")
        else:
            print("Invalid amount!")

    # Withdraw money
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"₹{amount} withdrawn successfully.")
        else:
            print("Insufficient Balance!")

    # Account info
    def account_info(self):
        print("\n--- Account Details ---")
        print("Account Holder:", self.name)
        print("Balance: ₹", self.balance)


# Create account
name = input("Enter Account Holder Name: ")
account = BankAccount(name, 1000)  # Initial balance

while True:
    print("\n===== BANK MENU =====")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Account Info")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        account.check_balance()

    elif choice == "2":
        amount = float(input("Enter amount to deposit: ₹"))
        account.deposit(amount)

    elif choice == "3":
        amount = float(input("Enter amount to withdraw: ₹"))
        account.withdraw(amount)

    elif choice == "4":
        account.account_info()

    elif choice == "5":
        print("Thank you for using Banking App!")
        break

    else:
        print("Invalid choice! Try again.")
```
