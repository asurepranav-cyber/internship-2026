```python
class Person:
    
    def __init__(self, name, age, ht):
        self.name = name
        self.age = age # private variable
        self.height = ht

    def show_info(self):
        return f"name -> {self.name}, age -> {self.age}, height -> {self.height}"
```

```python
p1 = Person('jatin', 20, 5.8)
p2 = Person('sachin', 22, 5.9)
```

```python
p1.show_info()
```

```python
"""
create a class called Employee =>
attributes -> company, name, monthly_salary, job_destination
here all the objects of this company will have the same common company
make salary as private variable


create 3 objects -> e1, e2, e3 with given attribute

method -> 1. method to calculate the annual salary
2. method to show entire data

""";
```

```python
class Employee:
    # Common company name for all employees (class variable)
    company = "Tech Solutions"

    def __init__(self, name, monthly_salary, job_destination):
        self.name = name
        self.__monthly_salary = monthly_salary   # private variable
        self.job_destination = job_destination

    # Method to calculate annual salary
    def calculate_annual_salary(self):
        return self.__monthly_salary * 12

    # Method to display employee details
    def show_data(self):
        print("Company:", Employee.company)
        print("Name:", self.name)
        print("Job Destination:", self.job_destination)
        print("Monthly Salary:", self.__monthly_salary)
        print("Annual Salary:", self.calculate_annual_salary())
        print("-" * 30)
```

```python
# Create 3 objects
e1 = Employee("Pranav", 50000, "Software Developer")
e2 = Employee("Amit", 60000, "Data Analyst")
e3 = Employee("Rahul", 70000, "AI Engineer")
```

```python
# Display data
e1.show_data()
e2.show_data()
e3.show_data()
```
