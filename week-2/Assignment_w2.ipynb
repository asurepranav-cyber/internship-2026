```python
from collections import Counter

def analyze_sentence(sentence):
    words = sentence.lower().split()

    result = {
        "word_count": len(words),
        "char_count": len(sentence.replace(" ", "")),
        "unique_words": len(set(words)),
        "most_frequent_word": Counter(words).most_common(1)[0][0]
    }

    return result


# Test
sentence = "The quick brown fox jumps over the lazy dog the fox"

output = analyze_sentence(sentence)
print(output)
```

Question 2

```python
text = "A man a plan a canal Panama"

# a) Reverse the string without slicing
reversed_text = ""
for ch in text:
    reversed_text = ch + reversed_text

print("Reversed:", reversed_text)


# b) Check palindrome (ignore spaces and case)
clean = text.replace(" ", "").lower()

if clean == "".join(reversed(clean)):
    print("Palindrome: Yes")
else:
    print("Palindrome: No")


# c) Replace every vowel with *
vowels = "aeiouAEIOU"
replaced = ""

for ch in text:
    if ch in vowels:
        replaced += "*"
    else:
        replaced += ch

print("Vowels replaced:", replaced)


# d) Count how many times each vowel appears
count = {"a": 0, "e": 0, "i": 0, "o": 0, "u": 0}

for ch in text.lower():
    if ch in count:
        count[ch] += 1

print("Vowel count:", count)
```

```python
def format_report(data):
    print(f"{'Name':<15}{'Score':>6}{'Grade':^8}")
    print("-" * 29)

    for name, score, grade in data:
        print(f"{name:<15}{score:>6}{grade:^8}")


data = [
    ("Alice", 92, "A"),
    ("Bob", 78, "B+"),
    ("Charlie", 55, "C"),
    ("Diana", 100, "A+")
]

format_report(data)
```

```python
def list_ops(nums):
    
    # a) Remove duplicates
    unique = []
    for num in nums:
        if num not in unique:
            unique.append(num)

    # b) Even indices squared
    squared = [nums[i]**2 for i in range(0, len(nums), 2)]

    # c) Flatten nested list
    nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
    flat = [item for sublist in nested for item in sublist]

    return unique, squared, flat


nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]

result = list_ops(nums)

print("Unique:", result[0])
print("Squared:", result[1])
print("Flattened:", result[2])
```

Question 7

```python
def inventory_manager(records):
    result = {}

    for product, category, quantity, price in records:

        if category not in result:
            result[category] = {
                "total_items": 0,
                "total_value": 0,
                "products": []
            }

        result[category]["total_items"] += quantity
        result[category]["total_value"] += quantity * price
        result[category]["products"].append(product)

    return result


records = [
    ("Laptop", "Electronics", 10, 55000),
    ("Mouse", "Electronics", 50, 499),
    ("Desk", "Furniture", 8, 12000),
    ("Chair", "Furniture", 15, 4500),
    ("Pen", "Stationery", 200, 20)
]

print(inventory_manager(records))
```

Question 10(b)

```python
import time


def timer(func):

    def wrapper(*args, **kwargs):

        start = time.time()

        result = func(*args, **kwargs)

        end = time.time()

        print("Execution Time:", end - start)

        return result

    return wrapper


@timer
def slow_sum(n):
    return sum(i*i for i in range(1, n+1))


print(slow_sum(100000))
```

Question 10(c)

```python
import math


def validate_positive(func):

    def wrapper(x):

        if x <= 0:
            raise ValueError(
                "Only positive numbers allowed"
            )

        return func(x)

    return wrapper


@validate_positive
def safe_sqrt(x):
    return math.sqrt(x)


print(safe_sqrt(25))
# print(safe_sqrt(-4))
```
