```python
def compare(a, b):
    if a > b:
        print(a, "is larger")
    elif b > a:
        print(b, "is larger")
    else:
        print("Both are equal")
```

```python
compare(10,20)
```

```python
compare(10,10) 
```

```python
#create a function to find the factorial of a number
```

```python
def factorial(n):
    fact = 1

    for i in range(1, n + 1):
        fact = fact * i

    print("Factorial =", fact)
```

```python
factorial(5)
```

```python
factorial(10)
```

```python
#function returning square and cube of a number
```

```python
def square_cube(n):
    square = n * n
    cube = n * n * n
    return square, cube
```

```python
s, c = square_cube(5)

print("Square =", s)
print("Cube =", c)
```

```python
#pass any number of integers in a function and find their sum
```

```python
def total(*num):
    s = 0

    for i in num:
        s = s + i

    print("Sum =", s)
```

```python
total(10, 20, 30, 40, 5, 1, 5)
```
