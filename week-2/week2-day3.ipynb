```python
"""
take a string as user input 
use a loop and if-else to check if a specific character is present in the string or not.
get output for both if the character is present and if it is not present.
take 2 characters to check in the string

user input -> string, character1, character2
""";
```

```python
sent = input("Enter a string: ")

char1 = input("Enter first character: ")
char2 = input("Enter second character: ")

present = False
for i in sent:
    if i == char1:
        present = True

if present:
    print(char1, "is present")
else:
    print(char1, "is not present")


present = False
for i in sent:
    if i == char2:
        present = True

if present:
    print(char2, "is present")
else:

    print(char2, "is not present")
```
