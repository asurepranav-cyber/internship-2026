Question 1: Take two numbers as input from the user and print their sum, difference, product, and quotient - each on a separate line. Handle division by zero with an appropriate message.

```python
num1 = float(input("Enter first number : "))
num2 = float(input("Enter second number: "))
 
print(f"\nSum        : {num1 + num2}")
print(f"Difference : {num1 - num2}")
print(f"Product    : {num1 * num2}")
 
if num2 == 0:
    print("Quotient   : Cannot divide by zero!")
else:
    print(f"Quotient   : {num1 / num2}")
 
```

Question 2: Temperature Conversion

```python
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9 / 5) + 32
kelvin = celsius + 273.15
 
print(f"\nCelsius    : {celsius} °C")
print(f"Fahrenheit : {fahrenheit} °F")
print(f"Kelvin     : {kelvin} K")
```

Question 4: Largest of Three Numbers

```python
a = int(input("Enter first number : "))
b = int(input("Enter second number: "))
c = int(input("Enter third number : "))
 
if a >= b and a >= c:
    largest = a
elif b >= a and b >= c:
    largest = b
else:
    largest = c
 
print(f"\nLargest number: {largest}")
```

Question 5: Leap Year Check

```python
year = int(input("Enter the Year"))
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("its a LEAP YEAR")
        else:
            print("not a leap year")
    else:
        print("It's a LEAP YEAR")
else:
    print("not a Leap Year")
```

Question 6: Simple Calculator

```python
x = float(input("Enter first number : "))
y = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ").strip()
 
if op == '+':
    print(f"\nResult: {x} + {y} = {x + y}")
elif op == '-':
    print(f"\nResult: {x} - {y} = {x - y}")
elif op == '*':
    print(f"\nResult: {x} * {y} = {x * y}")
elif op == '/':
    if y == 0:
        print("\nError: Cannot divide by zero!")
    else:
        print(f"\nResult: {x} / {y} = {x / y}")
else:
    print(f"\nError: '{op}' is not a valid operator!")
```

Question 8: Prime Numbers up to n

```python
limit = int(input("Enter a positive integer n: "))
 
print(f"\nPrime numbers from 2 to {limit}:")
primes = []
 
for num in range(2, limit + 1):
    is_prime = True
    for divisor in range(2, int(num ** 0.5) + 1):
        if num % divisor == 0:
            is_prime = False
            break
    if is_prime:
        primes.append(num)
 
if primes:
    print("  " + ", ".join(map(str, primes)))
else:
    print("  No prime numbers found in this range.")
```

Question 9: Running Sum & Average (stop at 0)

```python
numbers = []
 
while True:
    entry = float(input("Enter a number (0 to stop): "))
    if entry == 0:
        break
    numbers.append(entry)
 
if numbers:
    total = sum(numbers)
    count = len(numbers)
    average = total / count
    print(f"\nCount   : {count}")
    print(f"Sum     : {total}")
    print(f"Average : {average:.2f}")
else:
    print("\nNo numbers were entered.")
    
```

Question 10: Centered Triangle Pattern

```python
rows = int(input("Enter number of rows (n): "))
 
print()
for i in range(1, rows + 1):
    stars = 2 * i - 1           # stars in row i
    spaces = rows - i           # leading spaces for centering
    print(" " * spaces + "*" * stars)
```
