```python
#menu driven 
print("WELCOME TO THE Simple CALCULATOR")

num = int(input("ENTER THE NUMBER"))

print("Menu=>\n1. Sqauare\n2. Cube\n3. Check +ve or -ve")

ch = int(input("Select your Choice "))

if ch == 1:
    print(num * num)
elif ch == 2:
    print(num ** 3)
elif ch == 3:
    if num > 0:
        print("Positive")
    elif num == 0: 
        print("Zero")
    else:
        print("Negative")
else:
    print("Please enter the valid option /choice ")
```

```python

my_string = input("Enter a string: ")

while True:
    print("\n--- STRING MANIPULATION MENU ---")
    print("1. Convert to UPPERCASE")
    print("2. Convert to lowercase")
    print("3. Replace a part of the string")
    print("4. Exit")
    
    choice = input("Enter your choice (1/2/3/4): ")
    
 
    if choice == '1':
        print("\nResult:", my_string.upper())
        
    elif choice == '2':
        print("\nResult:", my_string.lower())
        
    elif choice == '3':

        target = input("Enter the part you want to replace: ")
        replacement = input(f"Enter what you want to replace '{target}' with: ")
        
        my_string = my_string.replace(target, replacement)
        print("\nResult:", my_string)
        
    elif choice == '4':
        print("\nExiting program. Goodbye!")
        break
        
    else:
        print("\nInvalid choice! Please enter 1, 2, 3, or 4.")
```
